```python
perceptrons_weights = []
for i in range(6): 
    perceptrons_weights.append(random.randrange(45,55))
lw1 = perceptrons_weights[:2]
lw2 = perceptrons_weights[2:4]
lw3 = perceptrons_weights[4:6]
wt = np.array([lw1,lw2,lw3]) 
wt
```

```python
def multiply(perceptrons_weights,sigarray):  
    mul = perceptrons_weights*sigarray
    return mul
```

```python
def sigmoid(x):
    z = 1/(1 + np.exp(-x))
    return z
vector = np.vectorize(sigmoid) 

for i in range(0,len(inp)):
    layer1 = vector(multiply(wt[0],inp[i])) #Values for layer1 (Input layer)
    layer2 = vector(multiply(wt[1],layer1)) #Values for layer2 (Hidden layer)
    layer3 = vector(multiply(wt[2],layer2)) #Values for layer3 (Output layer)
    
    #Backward Propagation
    
    er3 = (out[i] - layer3)*(np.ones(2) - layer3)*layer3*layer2 
    wt[2] = wt[2] + (alp * er3)
    er2 = (out[i] - layer3)*(np.ones(2) - layer3)* layer3 * wt[2] * (np.ones(2) - layer2) * layer2 * layer1 
    wt[1] = wt[1] + (alp * er2)
    er1 = (out[i] - layer3)*(np.ones(2) - layer3)*layer3*wt[2]*(np.ones(2) - layer2)*layer2*wt[1]*(np.ones(2) - layer1)*layer1*inp[i] 
    wt[0] = wt[0] + (alp * er1) 
print(wt)
```
